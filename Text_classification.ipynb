```python
import numpy as np
import pandas as pd
import os
import re
import nltk
from nltk.stem import WordNetLemmatizer 
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
data=pd.read_csv("/Users/user/Desktop/CUTE5/Data_Categories_TextClassification-1582933830366.csv") #Reading data
```

```python
data.head()
```

```python
Text_Description = data['Description']
```

# Text cleaning and preprocessing

```python
Text_Description_1 = Text_Description.tolist()
for i in range(len(Text_Description_1)):   
    Text_Description_1[i]=re.sub('[^A-Za-z]',' ',Text_Description_1[i])
    Text_Description_1[i]=re.sub('\s+',' ',Text_Description_1[i]) 
    #Replacing all the special characters other than alphabets with space using for loop
    #Also replacing more than one space with single space
```

```python
Text_Description[1]
```

```python
Text_Description_1[1]
```

```python
Text_Description_1=[token.lower() for token in Text_Description_1]
Text_Description_1[1]     #replacing the uppercase letters into lowercase letters
```

```python
for i in range(len(Text_Description_1)):
    Text_Description_1[i] = word_tokenize(Text_Description_1[i])
Text_Description_1[1]     #tokenizing or splitting all words in a sentence
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
print(stop)
```

```python
for i in range(len(Text_Description_1)):    #removing stowords present
    Text_Description_1[i] = [token for token in Text_Description_1[i] if token not in stop]
```

```python
Text_Description_1[1]    #we can see There is stopwords are removed now
```

# lemmatizaton

```python
lmtzr = nltk.stem.WordNetLemmatizer()
Text_Description_1[i] = [lmtzr.lemmatize(i) for i in Text_Description_1[i]]
print(Text_Description_1)
```

```python
for i in range(len(Text_Description_1)):
    Text_Description_1[i]=' '.join(Text_Description_1[i])
```

```python
Text_Description_1[1]
```

# Splitting data into train and test

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test=train_test_split(Text_Description_1,data['Category'],test_size=0.3,random_state=1234)
```

```python
#checking the distribution of target classes in train,here 3 classes 
y_train.value_counts()
```

```python
#checking the distribution of target classes in test,in test also there are same no of classes as in train
y_test.value_counts()
```

# Implementation of vector space models

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
tfidf_transformer = TfidfVectorizer(ngram_range=(1,1),stop_words='english')
X_train_tfidf = tfidf_transformer.fit_transform(X_train)
X_train_tfidf.shape
```

```python
X_train[0]
```

```python
X_test_tfidf = tfidf_transformer.transform(X_test)
X_test_tfidf.shape
```

# Development of classification models

# Implementing Naive Bayes Classification Model

```python
from sklearn.naive_bayes import MultinomialNB
NB = MultinomialNB().fit(X_train_tfidf, y_train) #NB is the classifier
pred_train = NB.predict(X_train_tfidf) #predictions on train data

from sklearn.metrics import confusion_matrix
prediction=confusion_matrix(y_train, pred_train) #confusion matrix for train predictions
```

```python
print("Train_Confusion Matrix:\n",prediction,"\n")

from sklearn.metrics import recall_score, precision_score, accuracy_score

acc = accuracy_score(y_train, pred_train)
rec = recall_score(y_train, pred_train, labels=[1,2,3],average='micro')
prec = precision_score(y_train, pred_train, labels=[1,2,3],average='micro')

print("Results of Naive Bayes on train data:","\nAccuracy:",acc, "\nRecall:",rec, "\nPrecision:",prec)
```

```python
pred = NB.predict(X_test_tfidf) #predicting on test data
prediction = confusion_matrix(y_test,pred) ##confusion matrix for test predictions
```

# Evaluation of model

```python
print("Test_Confusion matrix:\n", prediction,"\n")
acc = accuracy_score(y_test,pred)
rec = recall_score(y_test,pred,labels=[1,2,3],average='micro')
prec=precision_score(y_test,pred,labels=[1,2,3],average='micro')
print("Results of Naive Bayes on test data:","\nAcc: ",acc,"\nRec:", rec, "\nPrec:", prec)
```

# Implementation of Logistic Regression model

```python
from sklearn import linear_model

logreg_train = linear_model.LogisticRegression()
lr_clf = logreg_train.fit(X_train_tfidf, y_train) #fit logreg model on train data

pred_lr_train=lr_clf.predict(X_train_tfidf) # predict on train data
pred_lr_test=lr_clf.predict(X_test_tfidf) # predict on test data


print("Train_Confusion matrix:\n", confusion_matrix(y_train,pred_lr_train))
print("Test_Confusion matrix:\n",confusion_matrix(y_test,pred_lr_test))
```
